```python
# 15.11.2023
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(file_B3LYP, file_MP2):
  frames , electron, neutral = np.loadtxt(file_B3LYP, delimiter=',', unpack=True)
  VDE = []
  for index, frame in enumerate(frames):
    VDE.append((electron[index] - neutral[index]) * 27.211324570273)
  plt.plot(frames, VDE)

  frames , electron, neutral = np.loadtxt(file_MP2, delimiter=',', unpack=True)
  VDE = []
  for index, frame in enumerate(frames):
    VDE.append((electron[index] - neutral[index]) * 27.211324570273)
  plt.plot(frames, VDE)
  plt.ylim(-1,1.5)
  plt.grid()
```

```python
plot("ammonia-02-1-B3LYP.txt", "ammonia-02-1-MP2.txt")
```
