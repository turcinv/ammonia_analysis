```python
import matplotlib.pyplot as plt
from clusters.ammonia import Ammonia
import pandas as pd
import sqlite3
```

```python
def plot_average_distances(number):
  conn = sqlite3.connect('average_distance.db')

  query_nn = 'SELECT * FROM distances_nn WHERE cluster = ?'
  query_cn = 'SELECT * FROM distances_cn WHERE cluster = ?'

  data_NN = pd.read_sql_query(query_nn, conn, params=(number,))
  data_CN = pd.read_sql_query(query_cn, conn, params=(number,))
  
  conn.close()
  
  ammonia = Ammonia(number)
  print(ammonia)

  plt.figure(figsize=(12, 8))
  plt.title("Average distances", fontsize = 20)
  plt.xlabel("Time [fs]", fontsize = 15)
  plt.ylabel("Distance [A]", fontsize = 15)


  plt.plot(ammonia.data["time"], data_NN["distance"], label="Average distance NN")
  plt.plot(ammonia.data["time"], data_CN["distance"], label="Average distance CN")

  plt.grid()
  plt.legend(fontsize=15)
  plt.show()
```
